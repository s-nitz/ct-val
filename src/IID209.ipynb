```python
all_rows = test_df["taxonomy"].tolist()

counter = 0
for elt in all_rows:
    split_elt = elt.split(";")
    split_elt.insert(0, elt)
    
    if len(split_elt) != 8:
        nans_needed = 8 - len(split_elt)
        split_elt.extend(nans_needed * [0])
    
    new_df.loc[counter] = split_elt
            
    counter += 1
```
